### Split data

```python
# split the data set
from sklearn.model_selection import train_test_split

X = loan_df.drop('Personal_Loan', axis = 1)
y = loan_df['Personal_Loan']
```

### Check proportion of target data before balancing

```python
# check the proportion of the 2 classes in the personal loan
y.value_counts(normalize = True)
```

### Balance the data set

```python
# balance the dataset to get a better performance as the class we are interested in is very small
from imblearn.over_sampling import SMOTE

# instantiate the smote
smote = SMOTE(random_state=0)

X_sm, y_sm = smote.fit_resample(X, y)
```

### Check the proportion of balanced target data

```python
# check if the data is balanced
y_sm.value_counts(normalize = True)
```

```python
# split data into 80 and 20

x_train, x_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size= 0.2)
```

### Check to see proportion of split data

```python
# for training dataset
y_train.value_counts(normalize = True)
```

```python
# for test dataset
y_test.value_counts(normalize = True)
```

we see that both the training set and set test maintained the proportion of the categories of the target in both training and testing.

### Gradient Boost
### Using grid search to check for the best hyperparameter for gradient boost model

```python
# import GradientBoostingClassifier, the grid search model and the estimator
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV


# set the parameters to tune
param_grid = {'learning_rate': [0.001, 0.01, 0.1, 1, 10],
             'max_depth': list(range(1,5)),
             'criterion': ['friedman_mse', 'mse']}

# instantiate the models and grid search
gb = GradientBoostingClassifier(random_state=0)
grid = RandomizedSearchCV(gb, param_grid, cv = 5, n_jobs = -1, scoring = "recall")


# fit the grid search model on the training set
grid.fit(x_train, y_train)
```

```python
# so we want from the grid search is the best hyperparameter and best score from the hperparameter
best_params = grid.best_params_
best_params
```

now that we have this informatioin, we will use the best hyper param to build the model because we have not actually built any model

### Rebuild Gradient boost model with the best hyperparameter as indicated by the grid search

```python
gb2 = GradientBoostingClassifier(**best_params, random_state=0)

# train model with trainset
gb2.fit(x_train, y_train)
```

### check performance on train set

```python
# check model performance of the train set
gb_train = model_performance_class(gb2, x_train, y_train)
gb_train
```

On the train set, this model does not seem to overfitting but have a perfect recall score, lets evaluate the model against the test dataset

```python
gb_test = model_performance_class(gb2, x_test, y_test)
gb_test
```

with no misclassification of the assest customer and a perfect score of 100% to back it up, we know who uses personal loans and we will be sure that our assest customers are indeed assests and further actions to be taken on markerting will be directed on the other customers(liability) The gradient boost model have shown to be the best for this project.
